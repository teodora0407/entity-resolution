```python
import pandas as pd

df = pd.read_csv('entity_resolution.csv')
df.head(10)
```

```python
def clean_name(name):
    name = name.lower()

    for ch in ['.', ',', '-', '_']:
        name = name.replace(ch, '')

    name = name.strip()

    return name

df['cleaned_name'] = df['company_name'].apply(clean_name)

df.head(10)
```

```python
!pip install fuzzywuzzy[speedup]
```

```python
from fuzzywuzzy import fuzz

groups = []
used = set()

threshold = 90

for i in range(len(df)):
    if i in used:
        continue

    name1 = df.loc[i, 'cleaned_name']
    group = [df.loc[i, 'company_name']]

    for j in range(i + 1, len(df)):
        name2 = df.loc[j, 'cleaned_name']
        similarity = fuzz.token_set_ratio(name1, name2)

        if similarity >= threshold:
            group.append(df.loc[j, 'company_name'])
            used.add(j)

    used.add(i)
    groups.append(group)


for idx, group in enumerate(groups, 1):
    print(f"Grup {idx}: {group}")
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import numpy as np
from tqdm.notebook import tqdm

df = pd.read_parquet('veridion_entity_resolution_challenge.snappy.parquet')

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z0-9 ]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['cleaned_name'] = df['company_name'].apply(clean_text)

vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(3, 4))
tfidf_matrix = vectorizer.fit_transform(df['cleaned_name'])


n_neighbors = 20
threshold = 0.15

nn = NearestNeighbors(n_neighbors=n_neighbors, metric='cosine')
nn.fit(tfidf_matrix)

distances, indices = nn.kneighbors(tfidf_matrix)

group_labels = [-1] * len(df)
group_id = 0

for i in tqdm(range(len(df))):
    if group_labels[i] != -1:
        continue
    group_labels[i] = group_id
    for j in indices[i][1:]:
        if distances[i][list(indices[i]).index(j)] < threshold:
            if group_labels[j] == -1:
                group_labels[j] = group_id
    group_id += 1

df['group_id'] = group_labels


grouped = df[df.duplicated('group_id', keep=False)].groupby('group_id')['company_name'].apply(list)

for gid, names in grouped.head(10).items():
    print(f"\nGrup {gid}:")
    for name in names:
        print(" -", name)

```
